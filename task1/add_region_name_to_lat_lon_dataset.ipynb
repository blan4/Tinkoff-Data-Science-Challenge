```python
import pandas as pd
import numpy as np
```

```python
train_with_names = pd.read_csv('data/credit_train_clean.csv.zip', index_col='client_id', compression='gzip')
train_with_loc = pd.read_csv('data/credit_train_clean_loc.csv.zip', index_col='client_id', compression='gzip')

test_with_names = pd.read_csv('data/credit_test_clean.csv.zip', index_col='client_id', compression='gzip')
test_with_loc = pd.read_csv('data/credit_test_clean_loc.csv.zip', index_col='client_id', compression='gzip')
```

```python
all_with_names = pd.concat([train_with_names, test_with_names])
all_with_names.head()
```

```python
all_with_loc = pd.concat([train_with_loc, test_with_loc])
all_with_loc.head()
```

```python
all_data = pd.concat([all_with_loc, all_with_names['living_region']], axis=1)
print(all_data.columns)
all_data.head()
```

```python
all_data.ix[train_with_names.index].to_csv('data/credit_train_final.csv.zip', compression='gzip')
all_data.ix[test_with_names.index].drop(['open_account_flg'], axis = 1).to_csv('data/credit_test_final.csv.zip', compression='gzip')
```
